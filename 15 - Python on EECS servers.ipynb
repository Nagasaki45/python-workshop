### `rsync`
                                                                                               
Transfer files and directories either to or from a remote host.

- Transfer file from local to remote host:

```bash
rsync path/to/local_file username@remote_host:path/to/remote_directory
```

- Transfer file from remote host to local:

```bash
rsync username@remote_host:path/to/remote_file path/to/local_directory
```

- Transfer a directory. `-a` is for for **a**rchive: recursively, preserves attributes, and more. Check out the `--help`:

```bash
rsync -a path/to/directory username@remote_host:path/to/remote_directory
```

### `scp`

Stands for *Secure Copy*. Similar to the `cp` command. `rsync` is usually more useful.

### Jupyter notebooks over `ssh` tunnels

Both our local machine and the compute servers can connect to frank. Let's form a tunnel from our local machine, through frank, to the compute servers.

1. Run `jupyter notebook` on the compute server.

1. On the compute server create a reverse tunnel to frank:
```bash
ssh -R 9999:127.0.0.1:8888 username@frank.eecs.qmul.ac.uk
```
Now port 8888 on the compute serve is available as port 9999 on frank.

1. On your local machine create an ssh tunnel to frank:
```bash
ssh -L 8888:127.0.0.1:9999 username@frank.eecs.qmul.ac.uk
```
Now port 9999 on frank is available as port 8888 on our local machine.

Open the browser and go to [localhost:8888](http://localhost:8888).

**Note: ssh sessions on the compute servers (and the login servers) are killed after a few hours. Use `tmux` / `screen` to maintain the session.**
